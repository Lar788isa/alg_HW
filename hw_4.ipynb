Постройте график первой и второй производной функции (sin(2x+1))^5 на отрезке [-5; 5]

```python
# Посчитаем первую производную 
dif_1 = diff(pow(sin(2*x + 1),5))
dif_1
```

```python
#  Посчитаем вторую производную 
dif_2 = diff(10*pow(sin(2*x +1),4)*cos(2*x +1))
dif_2
```

```python
#  Закодируем исходную функцию 
def f(x):
    p = math.sin(2*x + 1)
    return pow(p,5)
```

```python
# Напишем код для 1-ой производной
def f1(x):
    return 10*pow(sin(2*x +1),4)*cos(2*x +1)
```

```python
def f2(x):
    return ((-20)*pow(math.sin(2*x +1),5) + 80*(pow(math.sin(2*x +1),3))*(pow(math.cos(2*x +1),2)))
```

```python
# посмотрим на графике как они выглядят
  
x1 = np.linspace(-5,5,100) # создали вектор х
y1 = [f(x1[i]) for i in range(len(x1))]
d1 = [f1(x1[i]) for i in range(len(x1))]  
d2 = [derivative(f1,x1[i],dx=1e-6,n=1) for i in range(len(x1))]
#d2 = [f2(x1[i]) for i in range(len(x1))]  

#plt.plot(x1,y1, label = 'функция')
plt.plot(x1,d1, label= '1-ая производная')  

plt.plot(x1,d2, label= '2-ая производная')

plt.title('Графики')
plt.ylabel('y')
plt.xlabel('x')

plt.legend()
plt.show()
```
