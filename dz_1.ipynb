Каждый дом - это вектор в 13-мерном пространстве. 
Написать код для подсчета скалярного произведения двух первых домов из датасета.

```python
x1 = boston.data[0] # дом 1
x2 = boston.data[1] # дом 2
```

```python
x1
```

```python
def skal(x1,x2):
    val= 0
    i = len(x1)-1
    while i >=0:
        val_0 = x1[i]* x2[i]
        val +=val_0
        i-=1
    return val
        
```

Найти косинус угла между векторами x1 и x2, где x1 и x2 - это два дома из 2ух атрибутов, полученных на этапе 2.

```python
#  Косинус угла между x1_2 и x2_2 равен скалярное произведение разделить на произведение длин двух векторов.
x1_Length = np.sqrt( (x1_2*x1_2).sum() ) # длина вектора
x2_Length = np.sqrt( (x2_2*x2_2).sum() )

cos_ = np.dot( x1_2, x2_2 ) / ( x1_Length * x2_Length )
cos_

# если значение косинуса близко к 1, то эти вектора близки. Это мы видим и на графике
```

```python

A = [
    [-100, -100, 1],
    [-100, 100, 1],
    [100, 100, 1],
    [0, 0, 1],
    [100, -100, 1],
    [-100, -100, 1]
    ]
A = np.array(A)
x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
# вторая матрица  - масштабирование на (0.5, 1,2)
    
TransformArray = np.array([
    [0.5, 0, 0],
    [0, 1.2, 0],
    [0,0,1]
    
])
```

```python
p = []
for point in A:
    new_point = np.dot(g, np.transpose(point))  # применяем первую матрицу
    new_point = np.dot(TransformArray, np.transpose(new_point))  #  применяем вторую матрицу
    p.append(new_point)
p = np.array(p).astype(int)
print(p)
```
