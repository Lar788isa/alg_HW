**Задание 1**
Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.


**Нормальное распределение**

```python
# Зададим параметры распределения 
mu = 2 #среднее значение 
sigma = 0.5 # квадрат величины отклонения от среднего значения 
sigma_1 = 0.2

# Зададим нормально распределенную случайную величину
norm_rv = sts.norm(loc=mu, scale = sigma)
norm_rv_1 = sts.norm(loc=mu, scale = sigma_1)
 
```

```python
# Сгенерируем 100 значений от 0 до 4
x=np.linspace(0,4,100)
pdf = norm_rv.pdf(x)
pdf1 = norm_rv_1.pdf(x)
plt.plot(x,pdf)
plt.plot(x,pdf1)

plt.plot(x,pdf, label= '1-ая')  

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```
