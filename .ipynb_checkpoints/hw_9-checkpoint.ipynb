```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

Создайте случайную величину из любого выбранного вами распределения, для разных значений n сгенерируйте 1000 выборок размера n и постройте гистрограммы средних этих выборок.

```python
# Зададим параметры нормального распределения 
mu = 2 #среднее значение 

sigma = 0.5 # квадрат величины отклонения от среднего значения (дисперссия)

rv = sts.norm(loc=mu, scale = sigma)
```

```python
sample = rv.rvs(size=1000)
```

```python
x = np.linspace(0,2,100)
pdf = rv.pdf(x)
plt.hist(sample, density=True, bins=30,label='sample histogram')
plt.plot(x, pdf, color = 'r',label='theoretical pdf', alpha=0.9)
plt.legend()
plt.ylabel('fraction of samples, $f(x)$')
plt.xlabel('$x$')
plt.show()
```

Выберем более наглядное значение mu

```python
# Зададим параметры нормального распределения 
mu = 1 #среднее значение 
sigma = 0.25 # квадрат величины отклонения от среднего значения 
rv = sts.norm(loc=mu, scale = sigma)
```

```python
sample = rv.rvs(size=1000)
```

```python
x = np.linspace(0,2,100)
pdf = rv.pdf(x)
plt.hist(sample, density=True, bins=30,label='sample histogram')
plt.plot(x, pdf, color = 'r',label='theoretical pdf', alpha=0.9)
plt.legend()
plt.ylabel('fraction of samples, $f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
#параметры запуска: размеры выборок и количество повторов генерации
n1 = 2
samples_count = 1000
#массивы для сохранения средних по выборкам
a1 = np.array([])

for number in range(samples_count):
    # на каждом цикле генерируем выборки разных размеров
    sample1 = rv.rvs(size=n1) #генерация выборки
    # считаем среднее по каждой выборке и добавляем в массив средних
    a1 = np.append(a1, sample1.mean())
    # YOUR CODE HERE
```

```python
#строим графики
#подписи для графиков
label_1 = 'sample of ' + str(n1) + ' histogram'
plt.hist(a1, density=True, label=label_1)
pdf = rv.pdf(x)
plt.plot(x, pdf, color = 'r',label='theoretical pdf', alpha=0.9)

plt.legend()
plt.ylabel('samples means')
plt.xlabel('$x$')
plt.show()
```

Поскольку значения дисперсии я не меняла , то построенная функция распределения плотности вероятности  поверх соответсвтующей гистограммы не  менялась. 

По полученным графикам наблюдается увелечение точности апроксимации выборочного среднего значения непрерывной случайной величины описываемой нормальным распределением. С ростом n точность апроксимации увеличивается.

А если мы дисперссию завяжим на n! 

```python
mu = 1 #среднее значение 
n3 = 100
sigma = mu/n3 #  чем больше значение n тем меньше дисперсия.

rv = sts.norm(loc=mu, scale = sigma)

for number in range(samples_count):
    sample1 = rv.rvs(size=n3) #генерация выборки
    a1 = np.append(a1, sample1.mean())
    
label_1 = 'sample of ' + str(n3) + ' histogram'
plt.hist(a1, density=True, label=label_1)
pdf = rv.pdf(x)
plt.plot(x, pdf, color = 'r',label='theoretical pdf', alpha=0.9)

plt.legend()
plt.ylabel('samples means')
plt.xlabel('$x$')
plt.show()
```

С уменьшение дисперсии, вероятность попасть в мат ожидание вырастает!
