**Задание 1**
Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.


**Нормальное распределение**

```python
# Зададим параметры распределения 
mu = 2 #среднее значение 

# квадрат величины отклонения от среднего значения на отрезке от 0,1 до 0,9 
sigma = np.linspace(0.1,0.9,5)
```

```python
# Сгенерируем 100 значений от 0 до 4
x=np.linspace(0,4,100)
```

```python
plt.rcParams["figure.figsize"] = (15,5) # растянем график для наглядности
```

```python
# Построим график плотности распределения случайных величин (при нормальном распределение)
for i in range(len(sigma)):
    norm_rv = sts.norm(loc=mu, scale = sigma[i])
    pdf = norm_rv.pdf(x)
    plt.plot(x,pdf, label= {f"sigma =", sigma[i]})
    plt.legend()

plt.ylabel('$плотность значений$')
plt.xlabel('$x$')

plt.show()
```

```python
# Рассмотрим разные значения - сред значения
mu1 = 2
mu2 = 3

sigma = 0.5 # квадрат величины отклонения от среднего значения 
sigma_1 = 0.25

# Зададим нормально распределенную случайную величину
norm_rv = sts.norm(loc=mu1, scale = sigma)
norm_rv_1 = sts.norm(loc=mu1, scale = sigma_1)

norm_rv_2 = sts.norm(loc=mu2, scale = sigma)
norm_rv_3 = sts.norm(loc=mu2, scale = sigma_1) 
```

```python
# Сгенерируем 100 значений от 0 до 4
x=np.linspace(0,4,100)
pdf = norm_rv.pdf(x)
pdf1 = norm_rv_1.pdf(x)

pdf2 = norm_rv_2.pdf(x)
pdf3 = norm_rv_3.pdf(x)

plt.plot(x,pdf, label= 'mu=2, sigma = 0.5')
plt.plot(x,pdf1, label='mu=2, sigma_1 = 0.25')

plt.plot(x,pdf2, label= 'mu=3, sigma = 0.5')
plt.plot(x,pdf3, label='mu=3, sigma_1 = 0.25')


plt.ylabel('$плотность значений$')
plt.xlabel('$x$')
plt.legend()
plt.show()
```

При нормальном распределение  изменение значения mu ведет к смещению графика , а изменения значения sigma влияет на плотность среднего значения , чем больше сигма тем ниже плотность в среднем значении. 

**Экспоненциальное распределение**

```python
# единственный параметр, полностью определяющий свойства данного распределения. Описывает среднее число наступлений события в ед времени 
lam =0.1
```

```python
expon_rv = sts.expon(lam) # экспоненциальное распределение
```

```python
x = np.linspace(-1,5,100) # Сгенерируем 100 значений от -1 до 5
pdf = expon_rv.pdf(x) # плотность вероятности 
```

```python
plt.plot(x,pdf)
plt.ylabel('$плотность значений$')
plt.xlabel('$x$')

plt.show()
```

```python
здесь плотность распределения может равняться нулю, когда значение СВ имеет отрицательное значение. 
```

```python
# Построим график плотности распределения случайных величин (при экспоненциальном распределение)
lam = np.linspace(0.1,0.9,5)
x = np.linspace(-1,5,100) # Сгенерируем 100 значений от -1 до 5

for i in range(len(lam)):
    expon_rv = sts.expon(lam[i]) # экспоненциальное распределение
    pdf = expon_rv.pdf(x) # плотность вероятности 
    
    plt.plot(x,pdf, label= {f"lam =", lam[i]})
    plt.legend()

plt.ylabel('$плотность значений$')
plt.xlabel('$x$')

plt.show()
```

Плотность распределения равна нулю при всех Х меньше значения lam

**Распределения Стьюдента непрерывных случайных величин**

```python
# Распределние Стьюдента симметрично. 
n =10
t_rv = sts.t(n) # распределние задаем 
```

```python
# Построим график плотности распределения Стьюдента случайных величин
n = np.linspace(0.9,10,5)
x = np.linspace(-10,10,100) # Сгенерируем 100 значений от -10 до 10

for i in range(len(n)):
    t_rv = sts.t(lam[i]) # экспоненциальное распределение
    pdf = t_rv.pdf(x) # плотность вероятности 
    
    plt.plot(x,pdf, label={f" n =",n[i]})
    plt.legend()

plt.ylabel('$плотность значений$')
plt.xlabel('$x$')

plt.show()
```

Чем больше степень свободы - значение n , тем компактнее  график распределения плотности.

**Задание 2**
Непрерывная случайная величина задана плотностью распределения:$$ f(\xi) = C, \xi \in [0;5];f(\xi) = 0, \xi \notin [0;5]$$Найдите C, математическое ожидание $\xi$ а также вероятность попадания $\xi$ в отрезок $[3.5;7]$, т.е. $P(3.5&lt;\xi&lt;7)$.


- У нас равномерное распределение непрерывной случайной величины , которая принимает значение  "С"  на отрезке от 0 до 5
- С = 1/ (в-а) = 1/ (5-0) = 0.2
-  вероятность попадания  𝜉  в отрезок  [3.5;7]. 

```python
a= 0
b=5
uniform_rv = sts.uniform(a, b-a)
```

```python
x= np.linspace(-1,10,100)
pdf = uniform_rv.pdf(x)
plt.plot(x,pdf)
plt.show()

# По графику видно, что значение С = 0,2!!
```

```python
cdf = uniform_rv.cdf(x)
plt.plot(x,cdf)
plt.show()
```
